### Data and Analysis for Section: [Global and Southern Ocean Heat and Carbon Content](#G_SO_OHC_OCC)  

###### Load Data

```python
PATHS={'aredi_800':'~/control_aredi800/data/newCO2_control_800/'}

for name, PATH in PATHS.iteritems():
    dic = iris.load_cube(PATH+'dic.nc')
    rhodzt = iris.load_cube(PATH+'rho_dzt.nc')
    area = iris.load_cube(PATH+'area_t.nc')
    
    if name == 'aredi_800':
        dic = dic[-500:]
```

###### Calculate heat and carbon content

```python
# Calculate Carbon Content: 
carbon, carbon_sumz, carbon_global = calc_occ(dic, rhodzt, area)
global_carbon_mean = carbon_global.collapsed('time', iris.analysis.MEAN)
global_carbon_anomaly = carbon_global-global_carbon_mean

# Calculate Heat Content:
heat, heat_sumz, heat_global = calc_ohc(temp, rhodzt, area)
global_heat_mean = heat_global.collapsed('time', iris.analysis.MEAN)
global_heat_anomaly = heat_global-global_heat_mean
```

###### Calculate Southern Ocean heat and carbon content

```python
# Southern Ocean Carbon Content:
constraint = iris.Constraint(latitude=lambda y: -90 < y < -50)
carbon_SO_tmp = carbon_sumz.extract(constraint)
area_SO = area.extract(constraint)
b = carbon_SO_tmp*area_SO
carbon_SO = b.collapsed(['longitude', 'latitude'], iris.analysis.SUM)

# Southern Oceam Carbon Content:
constraint = iris.Constraint(latitude=lambda y: -90 < y < -50)
heat_SO_tmp = heat_sumz.extract(constraint)
b = heat_SO_tmp*area_SO
heat_SO = b.collapsed(['longitude', 'latitude'], iris.analysis.SUM)

# Calculate Southern Ocean Carbon Content Anomoly
mean_SO = np.mean(carbon_SO.data)
carbon_SO_anomaly = carbon_SO.data - mean_SO

# Detrend Carbon with a polynomial
x = np.arange(0, 500, 1)
p = np.poly1d(np.polyfit(x, global_carbon_anomaly.data/1e15, 4))
p2 = np.poly1d(np.polyfit(x, carbon_SO_anomaly.data/1e15, 4))

global_carbon_detrend = global_carbon_anomaly.data/1e15 - p(x)
carbon_SO_detrend = carbon_SO_anomaly.data/1e15 - p2(x)

# Calculate Southern Ocean Heat Content Anomoly
mean_SO = np.mean(heat_SO.data)
heat_SO_anomaly = heat_SO.data - mean_SO
```

###### Isolate Convective Years

```python
## Use convective mask from temperature data and apply to heat and carbon content

# Carbon Content Masked
global_carbon_convect = ma.masked_array(global_carbon_detrend, convective_years.mask)
SO_carbon_convect = ma.masked_array(carbon_SO_detrend, convective_years.mask)

global_carbon_non = ma.masked_array(global_carbon_detrend, non_convective_years.mask)
SO_carbon_non = ma.masked_array(carbon_SO_detrend, non_convective_years.mask)

# Heat Content Masked
global_heat_convect = ma.masked_array(global_heat_anomaly.data, convective_years.mask)
SO_heat_convect = ma.masked_array(heat_SO_anomaly, convective_years.mask)

global_heat_non = ma.masked_array(global_heat_anomaly.data, non_convective_years.mask)
SO_heat_non = ma.masked_array(heat_SO_anomaly, non_convective_years.mask)
```

### Data and Analysis for Section: [Surface Heat Flux](#sfc_flux)

###### Load Data: 

```python
PATHS={'aredi_800':'/RESEARCH/hhpc/data/newCO2_control_800/'}

for name, PATH in PATHS.iteritems():
    sw_flx = iris.load_cube(PATH+'swflx.nc')
    lw_flx = iris.load_cube(PATH+'lw_heat.nc')
    sens_flx = iris.load_cube(PATH+'sens_heat.nc')
    evap_flx = iris.load_cube(PATH+'evap_heat.nc')
    area = iris.load_cube(PATH+'area_t.nc')
    dy_new = iris.load_cube(PATH+'dy.nc')
```

```python
# Zonally Average
sw_flx_zonalavg = sw_flx.collapsed('longitude', iris.analysis.MEAN)
lw_flx_zonalavg = lw_flx.collapsed('longitude', iris.analysis.MEAN)
sens_flx_zonalavg = sens_flx.collapsed('longitude', iris.analysis.MEAN)
evap_flx_zonalavg = evap_flx.collapsed('longitude', iris.analysis.MEAN)

# Time Mean
sw_flx_mean = sw_flx_zonalavg.collapsed('time', iris.analysis.MEAN)
lw_flx_mean = lw_flx_zonalavg.collapsed('time', iris.analysis.MEAN)
sens_flx_mean = sens_flx_zonalavg.collapsed('time', iris.analysis.MEAN)
evap_flx_mean = evap_flx_zonalavg.collapsed('time', iris.analysis.MEAN)
sum = sw_flx_mean + lw_flx_mean + sens_flx_mean + evap_flx_mean

# Zonally Integrate
sw_flx_W = sw_flx*area
lw_flx_W = lw_flx*area
sens_flx_W = sens_flx*area
evap_flx_W = evap_flx*area

sw_flx_integrate = sw_flx_W.collapsed('longitude', iris.analysis.SUM)
lw_flx_integrate = lw_flx_W.collapsed('longitude', iris.analysis.SUM)
sens_flx_integrate = sens_flx_W.collapsed('longitude', iris.analysis.SUM)
evap_flx_integrate = evap_flx_W.collapsed('longitude', iris.analysis.SUM)

sw_flx_integrate = (sw_flx_integrate/dy_new.data)*111131.7       # W/deg. latitude
lw_flx_integrate = (lw_flx_integrate/dy_new.data)*111131.7       # W/deg. latitude
sens_flx_integrate = (sens_flx_integrate/dy_new.data)*111131.7   # W/deg. latitude
evap_flx_integrate = (evap_flx_integrate/dy_new.data)*111131.7   # W/deg. latitude

total_heat_flux = sw_flx_integrate + lw_flx_integrate + sens_flx_integrate + evap_flx_integrate
```

###### Isolate convective and non-convective years

```python
## Convective:
convective_years_array = ma.masked_array(years, convective_years.mask)
convective_years_array = ma.compressed(convective_years_array)

sw_flx_convect = np.ones((len(convective_years_array), len(lats)))*np.nan
lw_flx_convect = np.ones((len(convective_years_array), len(lats)))*np.nan
sens_flx_convect = np.ones((len(convective_years_array), len(lats)))*np.nan
evap_flx_convect = np.ones((len(convective_years_array), len(lats)))*np.nan
total_convect = np.ones((len(convective_years_array), len(lats)))*np.nan


for idx, val in enumerate(convective_years_array):
    sw_flx_convect[idx,:] = sw_flx_integrate[val,:].data
    lw_flx_convect[idx,:] = lw_flx_integrate[val,:].data
    sens_flx_convect[idx,:] = sens_flx_integrate[val,:].data
    evap_flx_convect[idx,:] = evap_flx_integrate[val,:].data
    total_convect[idx,:] = total_heat_flux[val,:].data

sw_flx_convect_mean = np.mean(sw_flx_convect, 0)
lw_flx_convect_mean = np.mean(lw_flx_convect, 0)
sens_flx_convect_mean = np.mean(sens_flx_convect, 0)
evap_flx_convect_mean = np.mean(evap_flx_convect, 0)
total_convect_mean = np.mean(total_convect, 0)

## Non-convective:
non_convective_years_array = ma.masked_array(years, non_convective_years.mask)
non_convective_years_array = ma.compressed(non_convective_years_array)

sw_flx_non = np.ones((len(non_convective_years_array), len(lats)))*np.nan
lw_flx_non = np.ones((len(non_convective_years_array), len(lats)))*np.nan
sens_flx_non = np.ones((len(non_convective_years_array), len(lats)))*np.nan
evap_flx_non = np.ones((len(non_convective_years_array), len(lats)))*np.nan
total_non = np.ones((len(non_convective_years_array), len(lats)))*np.nan

for idx, val in enumerate(non_convective_years_array):
    sw_flx_non[idx,:] = sw_flx_integrate[val,:].data
    lw_flx_non[idx,:] = lw_flx_integrate[val,:].data
    sens_flx_non[idx,:] = sens_flx_integrate[val,:].data
    evap_flx_non[idx,:] = evap_flx_integrate[val,:].data
    total_non[idx,:] = total_heat_flux[val,:].data

sw_flx_non_mean = np.mean(sw_flx_non, 0)
lw_flx_non_mean = np.mean(lw_flx_non, 0)
sens_flx_non_mean = np.mean(sens_flx_non, 0)
evap_flx_non_mean = np.mean(evap_flx_non, 0)
total_non_mean = np.mean(total_non, 0)
```

# Define Convective Years <a id='Define_convection'></a>

#### Define convective years

Define convection to be any year when the Weddell Sea average temperature at 1400m is less than 1 standard deviation away from the mode. 

$$
t_{convect} = T_{1400m}(t) < \overline{T}_{1400m} - 2 \times s_{1400m}
$$

# Surface Heat Flux Analysis <a id='sfc_heat_flux'></a>

## Total Heat Flux <a id='total_flux'></a>

```python
# Total Heat
total_heat_convect = sw_flx_convect_mean + lw_flx_convect_mean + sens_flx_convect_mean + evap_flx_convect_mean
total_heat_non = sw_flx_non_mean + lw_flx_non_mean + sens_flx_non_mean + evap_flx_non_mean

total_heat_convect_mean = np.mean(total_heat_convect,0)
total_heat_convect_anomaly = total_heat_convect - total_heat_convect_mean

total_convect_mean2 = np.mean(total_convect_mean,0)
total_convect_anomoly = total_convect_mean - total_convect_mean2

fig3 = plt.figure(figsize = (8, 8))
ax2 = fig3.add_axes([0.15,0.55,0.7,0.35])
plt.title('(a) Total Heat Flux Anomoly', fontsize = 12)
plt.plot(lats, total_heat_convect_anomaly/1e12, color = 'r', lw = 1.5)
plt.plot(lats, total_heat_non_anomaly/1e12, color = 'b', lw = 1.5)
plt.axhline(0, color = 'k', ls = '--')
plt.ylabel(' Heat Flux [10$^{12}$ W deg$^{-1}$]', fontsize = 12)
plt.text(60, 201, 'Convective', color = 'r', fontsize = 10)
plt.text(60, 176, 'Non-convective', color = 'b', fontsize = 10)
plt.text(-75, 201, '* Positive Heats Ocean', color = 'k', fontsize = 10)


ax1 = fig3.add_axes([0.15,0.1,0.7,0.35])
plt.title('(b) Convective - Non-convective', fontsize = 12)
plt.plot(lats, (total_heat_convect-total_heat_non)/1e12, color = 'k', lw = 1.5)
plt.xlabel('Latitude', fontsize = 12)
plt.ylabel(' Heat Flux [10$^{12}$ W deg$^{-1}$]', fontsize = 12)
plt.axhline(0, color = 'k', ls = '--')
```

Strong ocean cooling in the high-latitude Southern Ocean during convection. Additionally, warming in the lower-latitude Southern Ocean, mid-latitudes, and tropics. 

## Individual Surface Fluxes <a id='individual_flux'></a>

```python
fig3 = plt.figure(figsize = (8, 8))
ax1 = fig3.add_axes([0.15,0.85,0.7,0.15])
ax1.plot(lats, (sw_flx_convect_mean - sw_flx_non_mean)/1e12, color = 'k', lw = 1.5)
plt.xlim([-80, 80])
plt.ylim([-12, 10])
plt.title('(a) Shortwave Heat Flux (convective - non-convective)', fontsize = 12)
plt.axhline(0, color = 'k', ls = '--')

ax1 = fig3.add_axes([0.15,0.6,0.7,0.15])
ax1.plot(lats, (lw_flx_convect_mean - lw_flx_non_mean)/1e12, color = 'k', lw = 1.5)
plt.xlim([-80, 80])
plt.ylim([-12, 10])
plt.title('(b) Longwave Heat Flux (convective - non-convective)', fontsize = 12)
plt.axhline(0, color = 'k', ls = '--')

ax1 = fig3.add_axes([0.15,0.35,0.7,0.15])
ax1.plot(lats, (sens_flx_convect_mean - sens_flx_non_mean)/1e12, color = 'k', lw = 1.5)
plt.xlim([-80, 80])
plt.ylim([-12, 10])
plt.title('(c) Sensible Heat Flux (convective - non-convective)', fontsize = 12)
plt.axhline(0, color = 'k', ls = '--')
plt.ylabel('                                              Heat Flux [10$^{12}$ W deg$^{-1}$]', fontsize = 12)


ax1 = fig3.add_axes([0.15,0.1,0.7,0.15])
ax1.plot(lats, (evap_flx_convect_mean - evap_flx_non_mean)/1e12, color = 'k', lw = 1.5)
plt.xlim([-80, 80])
plt.ylim([-12, 10])
plt.title('(d) Evaporative Heat Flux (convective - non-convective)', fontsize = 12)
plt.axhline(0, color = 'k', ls = '--')
plt.xlabel('Latitude')
```

Evaporative and longwave heat flux are primarily responsible for the loss of heat to the atmopshere in the high-latitude Southern Ocean (south of 60$^o$S). These fluxes balance offset the strong increase in shortwave heat flux in this region. In the lower-latitude Southern Ocean (40-60$^o$S), the increase in heat flux entering the icean seems to be due to an increase in the sensible heat flux, while the increases in the mid-latitudes are primarily due to shortwave.
