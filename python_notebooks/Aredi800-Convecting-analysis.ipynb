###### Import Modules

```python
import iris
import iris.analysis.cartography
import matplotlib.pyplot as plt
import numpy as np
import numpy as np
from scipy.stats import genextreme
import numpy.ma as ma
import seaborn as sns

import sys # access system routines
sys.path.append('/RESEARCH/hhpc/python')
from calc_region import calc_weddell_sea
import colormaps as cmaps
from calc_OHC_OCC import calc_occ, calc_ohc

%matplotlib inline 
```

### Data and Analysis for Section: [Convection Definition](#Define_convection) 

###### Load Data

```python
# Files loaded in this script were created in 'plot_aredi_climatologies.py' # 

PATHS={'aredi_800':'~/control_aredi800/data/newCO2_control_800/'}

for name, PATH in PATHS.iteritems():
    temp = iris.load_cube(PATH+'temp.nc')    
    temp_ws= calc_weddell_sea(temp)
    
    if name == 'aredi_800':
        temp = temp[-500:]
        
    temp_ws.coord('latitude').guess_bounds()
    temp_ws.coord('longitude').guess_bounds()
    grid_areas = iris.analysis.cartography.area_weights(temp_ws)
    temp_ws = temp_ws.collapsed(['latitude', 'longitude'], iris.analysis.MEAN, weights = grid_areas)
```

##### Define Convection

```python
mean = np.mean(temp_ws[:,18].data)
std = np.std(temp_ws[:,18].data)
convect = mean-(2*std) # Anything less than 2 standard deviation is considered to be convection
non_convect = mean-(std) # Anything less than 2 standard deviation is considered to be convection

time = np.arange(0, 500, 1)
convective_years = ma.masked_greater(temp_ws[:,18].data, convect)
non_convective_years = ma.masked_less_equal(temp_ws[:,18].data, non_convect)
```

### Data and Analysis for Section: [Global and Southern Ocean Heat and Carbon Content](#G_SO_OHC_OCC)  

###### Load Data

```python
PATHS={'aredi_800':'~/control_aredi800/data/newCO2_control_800/'}

for name, PATH in PATHS.iteritems():
    dic = iris.load_cube(PATH+'dic.nc')
    rhodzt = iris.load_cube(PATH+'rho_dzt.nc')
    area = iris.load_cube(PATH+'area_t.nc')
    
    if name == 'aredi_800':
        dic = dic[-500:]
```

###### Calculate heat and carbon content

```python
# Calculate Carbon Content: 
carbon, carbon_sumz, carbon_global = calc_occ(dic, rhodzt, area)
global_carbon_mean = carbon_global.collapsed('time', iris.analysis.MEAN)
global_carbon_anomaly = carbon_global-global_carbon_mean

# Calculate Heat Content:
heat, heat_sumz, heat_global = calc_ohc(temp, rhodzt, area)
global_heat_mean = heat_global.collapsed('time', iris.analysis.MEAN)
global_heat_anomaly = heat_global-global_heat_mean
```

###### Calculate Southern Ocean heat and carbon content

```python
# Southern Ocean Carbon Content:
constraint = iris.Constraint(latitude=lambda y: -90 < y < -50)
carbon_SO_tmp = carbon_sumz.extract(constraint)
area_SO = area.extract(constraint)
b = carbon_SO_tmp*area_SO
carbon_SO = b.collapsed(['longitude', 'latitude'], iris.analysis.SUM)

# Southern Oceam Carbon Content:
constraint = iris.Constraint(latitude=lambda y: -90 < y < -50)
heat_SO_tmp = heat_sumz.extract(constraint)
b = heat_SO_tmp*area_SO
heat_SO = b.collapsed(['longitude', 'latitude'], iris.analysis.SUM)

# Calculate Southern Ocean Carbon Content Anomoly
mean_SO = np.mean(carbon_SO.data)
carbon_SO_anomaly = carbon_SO.data - mean_SO

# Detrend Carbon with a polynomial
x = np.arange(0, 500, 1)
p = np.poly1d(np.polyfit(x, global_carbon_anomaly.data/1e15, 4))
p2 = np.poly1d(np.polyfit(x, carbon_SO_anomaly.data/1e15, 4))

global_carbon_detrend = global_carbon_anomaly.data/1e15 - p(x)
carbon_SO_detrend = carbon_SO_anomaly.data/1e15 - p2(x)

# Calculate Southern Ocean Heat Content Anomoly
mean_SO = np.mean(heat_SO.data)
heat_SO_anomaly = heat_SO.data - mean_SO
```

###### Isolate Convective Years

```python
## Use convective mask from temperature data and apply to heat and carbon content

# Carbon Content Masked
global_carbon_convect = ma.masked_array(global_carbon_detrend, convective_years.mask)
SO_carbon_convect = ma.masked_array(carbon_SO_detrend, convective_years.mask)

global_carbon_non = ma.masked_array(global_carbon_detrend, non_convective_years.mask)
SO_carbon_non = ma.masked_array(carbon_SO_detrend, non_convective_years.mask)

# Heat Content Masked
global_heat_convect = ma.masked_array(global_heat_anomaly.data, convective_years.mask)
SO_heat_convect = ma.masked_array(heat_SO_anomaly, convective_years.mask)

global_heat_non = ma.masked_array(global_heat_anomaly.data, non_convective_years.mask)
SO_heat_non = ma.masked_array(heat_SO_anomaly, non_convective_years.mask)
```

###### Analysis for Spatial Variations

```python
lats = carbon_sumz.coord('latitude').points
lons = carbon_sumz.coord('longitude').points
```

##### Use mask from subsurface temperature data to isolate convective and non-convective years

```python
years = np.arange(0,500,1)
convective_years_array = ma.masked_array(years, convective_years.mask)
convective_years_array = ma.compressed(convective_years_array)

heat_convect = np.ones((len(convective_years_array), len(lats), len(lons)))*np.nan
carbon_convect = np.ones((len(convective_years_array), len(lats), len(lons)))*np.nan

for idx, val in enumerate(convective_years_array):
    heat_convect[idx,:,:] = heat_sumz[val,:,:].data
    carbon_convect[idx,:,:] = carbon_sumz[val,:,:].data
heat_convect_mean = np.mean(heat_convect, 0)
carbon_convect_mean = np.mean(carbon_convect, 0)


# Non-convective
non_convective_years_array = ma.masked_array(years, non_convective_years.mask)
non_convective_years_array = ma.compressed(non_convective_years_array)

heat_non = np.ones((len(non_convective_years_array), len(lats), len(lons)))*np.nan
carbon_non = np.ones((len(non_convective_years_array), len(lats), len(lons)))*np.nan
for idx, val in enumerate(non_convective_years_array):
    carbon_non[idx,:,:] = carbon_sumz[val,:,:].data
    heat_non[idx,:,:] = heat_sumz[val,:,:].data
carbon_non_mean = np.mean(carbon_non, 0)
heat_non_mean = np.mean(heat_non, 0)

```

# Define Convective Years <a id='Define_convection'></a>

#### Define convective years

Define convection to be any year when the Weddell Sea average temperature at 1400m is less than 1 standard deviation away from the mean. 

$$
t_{convect} = T_{1400m}(t) < \overline{T}_{1400m} - 2 \times s_{1400m}
$$

```python
# Southern Ocean SST:
constraint = iris.Constraint(latitude=lambda y: -90 < y < -50)
sst_SO_tmp = sst.extract(constraint)
sst_SO_tmp.coord('latitude').guess_bounds()
sst_SO_tmp.coord('longitude').guess_bounds()
grid_areas = iris.analysis.cartography.area_weights(sst_SO_tmp)
sst_SO = sst_SO_tmp.collapsed(['longitude', 'latitude'], iris.analysis.MEAN, weights = grid_areas)

# Global SST: 
sst_global = sst.collapsed(['longitude', 'latitude'], iris.analysis.MEAN)


## Use convective mask from temperature data and apply to heat and carbon content

# Carbon Content Masked
global_sst_convect = ma.masked_array(sst_global.data, convective_years.mask)
SO_sst_convect = ma.masked_array(sst_SO.data, convective_years.mask)

global_sst_non = ma.masked_array(sst_global.data, non_convective_years.mask)
SO_sst_non = ma.masked_array(sst_SO.data, non_convective_years.mask)
```
